```python
# Imports 
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
sns.set_context('talk')

import warnings
warnings.filterwarnings('ignore')

import patsy
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest
# Note: the statsmodels import may print out a 'FutureWarning'. Thats fine.
```

```python
df1 = pd.read_csv('actors.csv')
grouped1 = df1.groupby('id')['name'].apply(list).reset_index()
grouped1.head(5)
```

```python
df5 = pd.read_csv('genres.csv')
grouped5 = df5.groupby('id')['genre'].apply(list).reset_index()
grouped5.head(5)
```

```python
df7 = pd.read_csv('countries.csv')
grouped7 = df7.groupby('id')['country'].apply(list).reset_index()
grouped7.head(5)
```

```python
df4 = pd.read_csv('movies.csv')
df4.rename(columns={'name': 'movie'}, inplace=True)
# Merge the two DataFrames on the 'id' column
merged_df14 = pd.merge(df4, grouped1, on='id', how='left')

```

```python
merged_df145 = pd.merge(merged_df14, grouped5, on='id', how='left')
```

```python
merged_df1457 = pd.merge(merged_df145, grouped7, on='id', how='left')
merged_df1457.head(5)
```

```python
df6 = pd.read_csv('crew.csv')
directors_df6 = df6[df6['role'] == 'Director']
directors_df6.rename(columns={'name': 'Director'}, inplace=True)
directors_df6.drop(columns='role', inplace=True)
directors_df6.head(5)
```

```python
merged_df14576 = pd.merge(merged_df1457, directors_df6, on='id', how='left')
merged_df14576.head(5)
```

```python
merged_df14576.drop(columns=['tagline', 'description', 'minute'], inplace=True)
merged_df14576.head(5)
```
